```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import ExtraTreesClassifier
```

```python
arq = pd.read_csv("C:/Users/jpsco/Documents/Professor/Doutorado/PGMC/BD2020/Dados/resultados/Banco35Exames.csv")
```

```python
# FUNES PARA TRANSFORMAR VARIVEIS LITERAIS EM NMERICAS

def transforma_estagio(valor):
    if   valor == 'Estgio 1 - >= 90 ml':  return 1
    elif valor == 'Estgio 2 - 60-89 ml':  return 2
    elif valor == 'Estgio 3a - 45-59 ml': return 3
    elif valor == 'Estgio 3b - 30-44 ml': return 4
    elif valor == 'Estgio 4 - 15-29 ml':  return 5
    elif valor == 'Estgio 5 - < 15 ml':   return 6

def transforma_raca(valor):
    if valor == 'Branca': return 0
    elif valor == 'Preta': return 1
    elif valor == 'Parda': return 2
    elif valor == 'Amarela': return 3
    elif valor == 'Indigena': return 4
```

```python
# CONVERSO DE VALORES LITERAIS PARA NUMRICOS

arq['ESTAGIOI - BIN'] = arq['ESTAGIOI - EQ'].map(transforma_estagio)
arq['ESTAGIOF - BIN'] = arq['ESTAGIOF - EQ'].map(transforma_estagio)

arq['Raa'] = arq['Raa'].map(transforma_raca)

arq['Codsexo'] = arq['Codsexo'].replace('Masculino', 0)
arq['Codsexo'] = arq['Codsexo'].replace('Feminino', 1)
```
